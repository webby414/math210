## II. Plotting with mplot3D

Lets plot the surface $z=x^2 + y^2$ for $x,y \in [-1,1]$

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
plt.figure() # creates a figure. window
ax = plt.axes(projection='3d')  # create a new axes 3d object in the current figure

x = np.linspace(-1,1,100)
y = np.linspace(-1,1,100)
X, Y = np.meshgrid(x,y)
Z = 1 - X**2 - Y**2
ax.plot_surface(X,Y,Z)
```

Lets plot $z=\cos(\sqrt(x^2 + y^2))$ from $x,y \in [-4\pi,4\pi]$

```python
plt.figure() # creates a figure. window
ax = plt.axes(projection='3d')  # create a new axes 3d object in the current figure

x = np.linspace(-8*np.pi,8*np.pi,500)
y = np.linspace(-8*np.pi,8*np.pi,500)
X, Y = np.meshgrid(x,y)
Z = np.cos(np.sqrt(X**2 + Y**2))
ax.plot_surface(X,Y,Z)
```

### The Meshgrid
Like many 3d plotting libraries (ie. MATLAB), the 3d plotting functions require 2d matrices of x y and z values. Mesh grid takes x and y arrays and returns two 2d matrices X and Y with X having the x values accross its rows and Y has the y values down its columns.

```python
x = [1,2,3]
y = [8,9,10,11]
```

```python
np.meshgrid(x,y)
```

The output of the meshgrid is a list with 2 entries. We can unpack those values into variables X and Y

```python
X, Y = np.meshgrid(x,y)
```

```python
plt.figure() # creates a figure. window
ax = plt.axes(projection='3d')  # create a new axes 3d object in the current figure

phi = np.linspace(0, np.pi,50)
thetha = np.linspace(0, 2*np.pi,50)

PHI, THETHA = np.meshgrid(phi,thetha)

X = np.sin(PHI) * np.cos(THETHA)
Y = np.sin(PHI) * np.sin(THETHA)
Z = np.cos(PHI)

ax.plot_wireframe(X,Y,Z)
plt.savefig('sphere.png', format='png')
```
