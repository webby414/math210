## I. Exploring `pyplot` Commands

We use the [matplotlib](http://matplotlib.org) to generate plots and we use the `pyplot` library of commands to customize our figure. There are many `pyplot` commands such as:
* `xlim` and `ylim`
* `grid`
* `title`

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Lets plot $y=\sin(x) + \sin(4x)$ in the range of $x \in [0,2\pi]$

```python
x = np.linspace(0,2*np.pi,10000)
y = np.sin(x) + np.sin(4*x)
plt.plot(x,y, 'r--', linewidth=2)
plt.xlim([0,2*np.pi]), plt.ylim([-3,3])
plt.grid('on', linestyle='-.')
plt.title('Sum of Sin Waves')
plt.xlabel('time (s)')
plt.ylabel('amplitude (mm)')
```

Theres a special command called `show`. We use this when we want to render the current figure. Jupyter notebook calls this for the current figure when it reaches the end of a cell

```python
x = np.linspace(0,2*np.pi,10000)
y1 = np.sin(x) + np.sin(4*x)
plt.plot(x,y1, '--', linewidth=2)
plt.show()

y2 = np.sin(x) + np.sin(4*x) + np.sin(4*x)
plt.plot(x,y2)
plt.show()
```

```python
t = np.linspace(0,4*np.pi,100)
y1 = np.sin(t)
y2 = np.cos(t)
plt.plot(t,y1,t,y2)
plt.legend(['y = sin(t), y = cos(t)'])
```
