## I. Plotting Examples

### Trefoil Knot in 2D

Given by the parametric equations 
$$x(t) = \sin(t) + 2\sin(2t) \\  \\ y(t)=\cos(t) - 2 \cos(2t)$$

```python
t = np.linspace(0,2*np.pi,1000)
x = np.sin(t) + 2*np.sin(2*t)
y = np.cos(t) - 2*np.cos(2*t)
```

```python
plt.plot(x,y)
plt.xlim([-4,4])
plt.axis('equal')
plt.ylim([-4,4])
plt.title('Trefoil Knot in 2D')
```

### Plotting Several Curves Together

Plotting the curves $y=e^{-x^2}\cos(kx)$ for $k = 1,2,5,10$ for $x \in [-5,5]$

```python
x = np.linspace(-5,5,1000)
y1 = np.exp(-x**2)*np.cos(x)
y2 = np.exp(-x**2)*np.cos(2*x)
y3 = np.exp(-x**2)*np.cos(5*x)
y4 = np.exp(-x**2)*np.cos(10*x)
plt.plot(x,y1,x,y2,x,y3,x,y4)
```

```python
x = np.linspace(-2,2,1000)

for k in [1,2,5,10]:
    y = np.exp(-x**2)*np.cos(k*x)
    plt.plot(x,y);
```

```python
x = np.linspace(-2,2,1000)
ks = np.random.rand(10)*10

for k in ks:
    y = np.exp(-x**2)*np.cos(k*x)
    plt.plot(x,y);
```

## II. Array Methods

NumPy arrrays have many methods (ie. functions) to do transformations and calculations. We acccess methods using the dot notation following the array variable name. For example,there is the `sum` method which simply sums the entries in the array.

```python
arr = np.array([1,2,3,4,5])
```

```python
arr.sum()
```

There is an equivulent way of doing this

```python
np.sum(arr)
```

Here we will take the product of the entry

```python
arr.prod()
```

The method cumsum computes the comulative sum of the entries:

```python
arr.cumsum()
```

```python
x = np.random.randn(1000,10).cumsum(axis=0)
# 1000 by 5 matrix. cumsum on the column
plt.plot(x);
```

We can also access the shape of the array using `.shape`

```python
M = np.random.randint(0,2,[4,5])
print(M)
M.shape
```

Notice that we accessed the shape without using parenthese. This is the differences between array methods (functions to do calculations with the array -- like it's shape).

```python
M.sum()
```

```python
M.shape
```

There is also the transpose method:

```python
a = np.array([[1,2],[3,4]])
a
```

```python
a.transpose()
```

```python
a.mean()
```

```python
a.mean(axis=0) # axis 0 is accross the rows
```

```python
a.mean(axis=1) # axis 1 is down the collumns
```

## III. Array Slicing and Indexing

We access the entries in an array using the square brackets

```python
A = np.random.randint(0,10,[3,5])
A
```

```python
A[2,4] # row index 2 and column index 4
```

```python
A[0,0]
```

To access a whole role (or column) we use the colon : 

```python
A[0,:]
```

```python
A[:,3]
```
