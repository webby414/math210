## More 3D Plotting

### A Saddle
Plotting now $z = x^2 - y^2$ for $x,y \in [-2,2]$

```python
plt.figure()
ax = plt.axes(projection="3d")

x = np.linspace(-2,2,100) 
y = np.linspace(-2,2,100) 
X, Y = np.meshgrid(x,y)
Z = X**2 - Y**2
ax.plot_surface(X,Y,Z)
```

```python
plt.figure()
ax = plt.axes(projection="3d")

x = np.linspace(-2,2,50) 
y = np.linspace(-2,2,50) 
X, Y = np.meshgrid(x,y)
Z = X**2 - Y**2
ax.plot_wireframe(X,Y,Z)
```

### Plotting the trefoil knot in 3D
$$
x = \sin(t) +2\sin(2t) \ \ , \ \ y = \cos(t) -2\cos(t) \ \ , \ \ z = -3\sin(t)
$$

For $t \in [0,2\pi]$

```python
plt.figure()
ax = plt.axes(projection='3d')

t = np.linspace(0,2*np.pi, 100)

x = np.sin(t)+2*np.sin(2*t)
y = np.cos(t)-2*np.cos(2*t)
z = -np.sin(3*t)

ax.view_init(10,10)
ax.plot(x,y,z)
```

### The Taurus

```python
plt.figure()
ax = plt.axes(projection='3d')

R = 3
r = 1

theta = np.linspace(0,2*np.pi,50)
phi = np.linspace(0,2*np.pi,50)

PHI, THETHA = np.meshgrid(phi,theta)

X = (R + r*np.cos(THETHA))*np.cos(PHI)
Y = (R + r*np.cos(THETHA))*np.sin(PHI)
Z = r*np.sin(THETHA)

ax.view_init(60,30)
ax.plot_wireframe(X,Y,Z)
```

### Keyword Arguments
we can define functions with keyword argumetns such as:

```python
def fun(a,b,add=True):
    
        if add == True:
            return a + b
        else:
            return a*b
```

```python
fun(1,2)
```

### List comprehensions with if/else

Let's make a list of integers from 1 to 100 which are multiples of 3 or 7

```python
a = [n for n in range(1,101) if(n % 3 == 0 or n % 7 == 0)]
print(a)
```
