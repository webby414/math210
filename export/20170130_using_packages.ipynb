## I. Creating NumPy Arrays

The main features of NumPy are: 
1. `ndarray`: Numpy's $n$-dimensional array object
2. Vectorized mathematical functions which broadcast across arrays for fast computation

To get started with NumPy, let's import it (under an alias np) - standard convention

```python
import numpy as np
```

Follow an object or function name with a question mark and the Jupyter notebook will display its documentation.

```python
np?
```

Enter `np.<TAB>` (`np.` followed by the <TAB> jey) to see a list of what NumPy contains.

Use the function `np.array` to create a NumPy array from a Python list (or tuple):

```python
a = np.array([1,2,4])
print(a)
```

```python
type(a)
```

```python
a.shape
```

Create a 2D array (ie. matrix) from a Python list of lists:

```python
M = np.array([[1,2,3],[4,5,6]])
print(M)
```

We can make a n-dimensional array from a nested Python list. For example, the following is a 3D array

```python
N = np.array([[[1,2],[3,4]],[[5,6],[7,8]]])
print(N)
```

## II. Random Number Generator

The subpackage `numpy.random` contains functiosn to generate random numbers from different distributions. Enter `np.random.<TAB>` (`np.` followed by the <TAB> jey) to see a list of what NumPy Random contains and use the question marks to see the documentation.

```python
np.random.rand?
```

```python
N = 1000
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
sizes = np.random.rand(N)*100
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(x,y,c=colors,s=sizes,lw=0,alpha=0.2)
```

```python
np.random.randn()
```

```python
N = 4000
x = np.random.randn(N)
y = np.random.randn(N)
plt.scatter(x,y,lw=0,alpha=0.2)
```

```python
np.random.randint(0,7)
```

```python
roll = np.random.randint(0,7)

while roll != 6:
    print(roll)
    roll = np.random.randint(0,7)
```
