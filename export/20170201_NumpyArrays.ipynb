## I. NumPy Arrays

### Array operations
We can do arithmetic operations on arrays like addition, multiplication and division which are performed element-wise.

Array addition/subtraction is the same as vector addition/subtraction.

```python
import numpy as np
```

```python
v = np.array([1,2,3])
w = np.array([1,0,-1])
```

```python
v + w
```

```python
v - w
```

In the same way, array multiplication/division is performed element by element

```python
v*w
```

```python
w/v
```

Same with exponents

```python
v**w
# the solution got floored. It turns out that numpy arrays are homogenou, since they only 
# carry one type of data. Because we defined these arrays with integer entries and most of the 
# entries in `v**w` are integers, the array remains an integer arrays
```

```python
v ** 2 
```

```python
v.dtype # the entries of v are 64bit integers
```

```python
(1/v).dtype # the entries of 1/v are floats
```

```python
v = np.arange(0,1,.1)
```

```python
v
```

```python
w = np.arange(1,11)
```

```python
v ** w
```

### Vectorized mathematical functions
A vector function is just a function which gets applied to every entry of an array

```python
np.pi
```

```python
x = np.array([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
print(x)
```

```python
np.sin(x)
```

```python
x = np.linspace(-50,50,1000)
y = np.sin(x)
plt.plot(x,y), plt.ylim([-2,2]), plt.xlim([-50,50]);
```

let plot $y=e^{x^2}\cos(10x)$

```python
x = np.arange(-10,10,0.01)
y = np.exp(-x**2/10) * np.cos(10*x)
plt.plot(x,y), plt.ylim([-2,2])
```

Let's plot $x = \cos(t)$ and for $\sin(t)$ for $t \in [0,2\pi]$ 

```python
t = np.linspace(0,2*np.pi,1000)
x = t*np.cos(10*t)
y = t*np.sin(10*t)
plt.plot(x,y), plt.axis('equal'), plt.xlim([-2,2]), plt.ylim([-2,2])
```
